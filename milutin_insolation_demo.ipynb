```python
import milutin as mt
```

This notebook demonstrates the usage of the `Insolation()` function from the Milutin Python package, which calculates the top-of-atmosphere (TOA) insolation for different latitudes, seasons, and years BP. 

## Function Overview

milutin.insolation(kyear = None, latitude = None, output_type = 'array', show_plot = True, season = None, day = None, filename = 'milutin_plot')

Parameters:

- `kyear` optional, unit: kyr, type:(int,float,tuple,list or np.ndarray) - If kyear is a int or float, return TOA insolation at that kyear. If kyear is tuple of length 3, items will be interpreted as the respective orbital parameters: eccentricity, obliquity, and longitude of perihleion, return TOA insolation for parameter set. If kyear is tuple of length 2, items will be interpreted as domain boundries of kyear range, return insolation array over domain range. If kyear is 1-D list or np.ndarray, returns insolation values at each item value in array. If not specified, returns modern insolation.

- `latitude` optional, unit: degrees, type:(int,float,tuple,list or np.ndarray) - If latitude is int or float, return TOA insolation at that latitude. If latitude is tuple of length 2, items will be interpreted as domain boundries over latitude range, return TOA insolation over latitude range. If latitude is 1-D array like, returns TOA insolation values at each latitude value in array. If not specified, returns TOA insolation over range (-90,90).

- `season` optional, type:(string) - Specifies the time domain over which to solve TOA insolation, accepted strings are, boreal winter/austral summer: 'DJF', boreal spring/austral autumn: 'MAM', boreal summer/austral winter: 'JJA', and boreal autumn/austral spring: 'SON'. If not specifed, insolation time domain will refer to day argument. 

- `days` optional, unit: days of the year type:(tuple): Allows manual selection of days of the year time domain over which to solve TOA insolation. a value of 0 is the 1st of January, and 365 is the 31st of December. If day is tuple of length 2, items will be interpreted as domain boundries over day of the year range. If not specified, returns TOA insolation over entire year.

- `output_type`: optional, type:(string) - Specifies which dimention of output to average over. Valid inputs vary based on other input parameters but arguments include: "latitude mean" to average over latitudes, "day mean" to average over days of the year, "kyear mean" to average over kyears, "latitude day mean" to average over days and latitudes if day and/or latitude range does not span entire valid domain, and "global annual mean" to average over days and latitudes if day and/or latitude range does span entire valid domain. If not specified, no dimentional averages will be determined in output.

- `filename`: optional, type:(string) - File or filename for which the figure is saved. If not specified, figure will be saved as "milutin_plot.png".

- `show_plot`: optional, type:(bool) - Allow saving output figure to file. If not specified output will save to filename.

Returns:

- TOA Insolation: xarray.DataArray

```python
# default insolation run
mt.insolation()
```

# Kyear Argument Examples

```python
# use int or float to specify a kyear BP to solve TOA insolation for
mt.insolation(kyear=21.)
```

```python
# use tuple of length 2 to solve TOA insolation over kyear range:
mt.insolation(kyear = (10,0))
```

```python
# use 1-D array like object such as list or np.ndarray to solve for specific kyears
kyear_array = [0, 21, 120, 1200, 3000, 7000]
mt.insolation(kyear = kyear_array)
```

```python
# use tuple of length 3 to solve TOA insolation for specific orbital parameters
mt.insolation(kyear = (0.001,50,300))
```

# Latitude Argument Examples

```python
# use int or float to specify which latitudes to solve TOA insolation for:
mt.insolation(latitude = 65)
```

```python
# use 1-D array like object to solve for multiple latitudes 
latitude_list = [-65,-10.5,0,10.5,65]
mt.insolation(latitude = latitude_list)
```

```python
# use tuple of length 2 to specify range of latitudes to solve TOA insolation for
mt.insolation(latitude = (-10,10))
```

# Season Argument Examples

```python
# specify boreal winter TOA insolation
mt.insolation(season = 'DJF')
```

```python
# specify boreal summer insolation
mt.insolation(season='JJA')
```

# Day Argument Example

```python
# use days tuple to specify range of days
mt.insolation(days = (0,100))
```

# Output_type Argument Exmaples

```python
# use output_type 'global annual mean' over range of kyears:
mt.insolation(kyear = (3000,0),output_type = 'global annual mean')
```

```python
# find daily mean insolation
mt.insolation(output_type = 'day mean')
```

```python
# find the average insolation field for a range of kyears
mt.insolation(kyear = (10,0), output_type='kyear mean')
```

```python
# find insolation at 65N on June 21st over kyear range
mt.insolation(kyear = (150,0), latitude = 65, days = (171,172), output_type='day mean')
```

```python
# We can look at insolation from 130kyr BP to modern over the Southern Ocean averaged by day and latitude
mt.insolation(kyear = (130,0), latitude = (-75,-45), output_type='latitude day mean')
```

```python
mt.insolation(kyear=(200,0),latitude=[-45,45],output_type='day mean')
```
